```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('_', ' ', regex=True)
df.columns = df.columns.str.title()
df.columns
```

```python
df['Show Id'] =df['Show Id'].str.replace('s','',regex=True)
df['Show Id'] = df['Show Id'].astype(int)
df['Show Id']
```

```python
df = df.sort_values(by='Show Id', ascending=True)
df = df.reset_index(drop=True)
df.head()
```

```python
df['Date Added'] = df['Date Added'].str.strip()
```

```python
df['Date Added'] = df['Date Added'].astype('datetime64[ns]')
```

```python
df.dtypes
```

```python
df.drop(columns='Rating', inplace=True)
df.drop(columns='Duration', inplace=True)
df.head()
```
