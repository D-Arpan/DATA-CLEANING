```python
columns_to_keep = ['asin', 'product_title', 'product_price', 'product_star_rating','product_num_ratings', 'product_num_offers', 'product_minimum_offer_price', 'is_best_seller',
                   'is_amazon_choice', 'is_prime', 'climate_pledge_friendly','sales_volume', 'has_variations']
columns_to_drop = ['Unnamed: 0', 'product_original_price', 'product_availability', 'unit_price', 'unit_count', 'currency', 'product_url', 'product_photo', 'delivery']
```

```python
df=df[columns_to_keep]
```

```python
df.columns = df.columns.str.upper().str.replace('_',' ',regex=True).str.strip()
```

```python
df.head()
```

```python
df.rename(columns={'PRODUCT PRICE':'PRODUCT PRICE (USD)'}, inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['PRODUCT STAR RATING'] = df['PRODUCT STAR RATING'].fillna(df['PRODUCT STAR RATING'].mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['PRODUCT PRICE (USD)'] = df['PRODUCT PRICE (USD)'].str.lstrip('$').str.strip()
```

```python
df['PRODUCT PRICE (USD)'] = df['PRODUCT PRICE (USD)'].astype(float)
df.head()
```

```python
df['PRODUCT MINIMUM OFFER PRICE'] = df['PRODUCT MINIMUM OFFER PRICE'].str.lstrip('$').str.strip()
df['PRODUCT MINIMUM OFFER PRICE'] = df['PRODUCT MINIMUM OFFER PRICE'].astype(float)
df.head()
```

```python
df.replace(True,1,inplace=True)
df.replace(False,0,inplace=True)
```

```python
df['PRODUCT MINIMUM OFFER PRICE']=df['PRODUCT MINIMUM OFFER PRICE'].round()
df['PRODUCT PRICE (USD)']=df['PRODUCT PRICE (USD)'].round()
```

```python
df['PRODUCT PRICE (USD)'] = df['PRODUCT PRICE (USD)'].astype(int)
df['PRODUCT MINIMUM OFFER PRICE']=df['PRODUCT MINIMUM OFFER PRICE'].astype(int)
```

```python
df['PRODUCT STAR RATING'] = df['PRODUCT STAR RATING'].round(1)
```

```python
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.to_excel('amazon_product_cleaned.xlsx',index=False)
df.to_csv('amazon_product_cleaned_csv.csv',index=False)
```
