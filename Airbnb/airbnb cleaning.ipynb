## DELETING REDUNDANT COLUMNS

```python
df.drop(['last review', 'reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license'], axis=1 , inplace=True)
```

```python
df.shape
```

## RENAMING COLUMNS

```python
new_columns_name = []
for i in df.columns:
    new_columns_name.append(i.upper())
```

```python
new_columns_name
```

```python
df.columns=new_columns_name
df.columns
```

```python
df.columns = df.columns.str.replace('_',' ')
df.columns
```

```python
df.duplicated().sum()
```

## REMOVING DUPLICATES

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

## REMOVING NaN OR N/a VALUES

```python
df.isna().sum()
```

```python
df.dropna(subset=['NAME'], inplace=True)
df.dropna(subset=['HOST IDENTITY VERIFIED'], inplace=True)
df.dropna(subset=['HOST NAME'], inplace=True)
df.dropna(subset=['NEIGHBOURHOOD GROUP'], inplace=True)
df.dropna(subset=['NEIGHBOURHOOD'], inplace=True)
df.dropna(subset=['COUNTRY'], inplace=True)
df.dropna(subset=['LAT','LONG','COUNTRY CODE','INSTANT BOOKABLE','CONSTRUCTION YEAR'], inplace=True)
```

```python
df.isna().sum()
```

## FILLING NaN VALUES WITH SUITABLE VALUES

```python
df.dtypes
```

```python
# THE PRICE HAD SYMBOLS LIKE $966 (DOLLAR) AND $1,060 (COMA)
df['PRICE'] = df['PRICE'].str.replace('[^0-9]', '', regex=True)
```

```python
# CHANGING THE DATATYPE TO PERFORM STATISTICAL ACTIONS
df['PRICE'] = df['PRICE'].astype(float)
```

```python
df.info()
```

```python
df['PRICE'].fillna(df['PRICE'].mean(), inplace=True)
```

```python
df.drop(columns='NUMBER OF REVIEWS', inplace=True)
```

```python
df['SERVICE FEE'] = df['SERVICE FEE'].str.replace('[^0-9]','', regex=True)
```

```python
df.info()
```

```python
df['SERVICE FEE'] = df['SERVICE FEE'].astype(float)
```

```python
df.dtypes
```

```python
df['SERVICE FEE'].fillna(df['SERVICE FEE'].mean(), inplace=True)
```

```python
df.info()
```

```python
df['MINIMUM NIGHTS'] =  df['MINIMUM NIGHTS'].fillna(df['MINIMUM NIGHTS'].mode().iloc[0])
```

```python
df['MINIMUM NIGHTS'] = df['MINIMUM NIGHTS'].astype(int)
```
